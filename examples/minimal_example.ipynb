## PYORPS Minimal Example: Optimal Route Planning for Underground Cables
This minimal example demonstrates how to use PYORPS to determine cost-effective and environmentally compatible routes for underground cables. PYORPS performs least-cost path analysis on high-resolution rasterized geodata, making it particularly suitable for planning tasks in distribution systems. This example will guide you in effectively utilizing PYORPS, showcasing how it can assist you in making informed decisions to enhance the planning process for underground cable installations.

#### 1. Import PathFinder from PYORPS
Begin by importing the PathFinder class from the PYORPS library, which will be used to create a graph from the raster data and to perform the least-cost path analysis.

```python
from pyorps import PathFinder
```

#### 2. Define Source and Target Coordinates
Specify the source and target points for the route by providing their coordinates in the coordinate reference system
(CRS) that corresponds to the rasterized geodata.

```python
source = (472000, 5593400)
target = (472800, 5594000)
```

#### 3. Create a PathFinder Instance
Instantiate the PathFinder class by passing the defined source and target coordinates along with a data source, which can be a local raster dataset — one of several options for conducting the analysis. If no suitable raster dataset is available, guidance on creating a raster by defining terrain-specific cost assessments is provided in the "create_rasterized_geodata.ipynb" notebook.

```python
path_finder = PathFinder(
    dataset_source=r"data\raster\sample_raster.tiff",
    source_coords=source,
    target_coords=target,
)
```

#### 4. Run the Least-Cost Path Analysis
 Initiate the pathfinding process by calling the find_route method on the PathFinder instance. This method will
 return a Path object, which contains the optimal route and can be further analyzed.

```python
path_finder.find_route()
```

#### 5. Plot the Path on the Raster Data
Visualize the resulting path by plotting it over the raster dataset, allowing
for a clear representation of the optimal route determined by the analysis.

```python
path_finder.plot_paths()
```

#### 6. Save the Path Data as a GeoJSON File
Export the path data to a GeoJSON file format, enabling its use in GIS  applications for further investigation and adjustments according to user needs.

```python
path_finder.save_paths(r"data\results\route_from_minimal_example.geojson")
```
